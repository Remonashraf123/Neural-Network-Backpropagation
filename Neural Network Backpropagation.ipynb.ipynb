```python
import numpy as np

# Activation Function (Sigmoid)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of Sigmoid
def sigmoid_derivative(x):
    return x * (1 - x)

# Inputs and Outputs
inputs = np.array([[0.05, 0.1]])
targets = np.array([[0.01, 0.99]])

# Weights
w1 = 0.15
w2 = 0.20
w3 = 0.25
w4 = 0.30
w5 = 0.40
w6 = 0.45
w7 = 0.50
w8 = 0.55
b1 = 0.35
b2 = 0.60

# Learning Rate
eta = 0.5

# Forward Pass (Hidden Layer)
net_h1 = inputs[0, 0] * w1 + inputs[0, 1] * w3 + b1
out_h1 = sigmoid(net_h1)

net_h2 = inputs[0, 0] * w2 + inputs[0, 1] * w4 + b1
out_h2 = sigmoid(net_h2)

# Forward Pass (Output Layer)
net_o1 = out_h1 * w5 + out_h2 * w7 + b2
out_o1 = sigmoid(net_o1)

net_o2 = out_h1 * w6 + out_h2 * w8 + b2
out_o2 = sigmoid(net_o2)

# Error Calculation
E1 = 0.5 * (targets[0, 0] - out_o1) ** 2
E2 = 0.5 * (targets[0, 1] - out_o2) ** 2
E_total = E1 + E2

# Backpropagation
# Output Layer
# For O1
dE_total_out_o1 = out_o1 - targets[0, 0]
dE_total_net_o1 = dE_total_out_o1 * sigmoid_derivative(out_o1)

# For O2
dE_total_out_o2 = out_o2 - targets[0, 1]
dE_total_net_o2 = dE_total_out_o2 * sigmoid_derivative(out_o2)

# Hidden Layer
# For H1
dE_total_out_h1 = (dE_total_net_o1 * w5) + (dE_total_net_o2 * w6)
dE_total_net_h1 = dE_total_out_h1 * sigmoid_derivative(out_h1)

# For H2
dE_total_out_h2 = (dE_total_net_o1 * w7) + (dE_total_net_o2 * w8)
dE_total_net_h2 = dE_total_out_h2 * sigmoid_derivative(out_h2)

# Weight
w1 = w1 - eta * dE_total_net_h1 * inputs[0, 0]
w2 = w2 - eta * dE_total_net_h2 * inputs[0, 0]
w3 = w3 - eta * dE_total_net_h1 * inputs[0, 1]
w4 = w4 - eta * dE_total_net_h2 * inputs[0, 1]
w5 = w5 - eta * dE_total_net_o1 * out_h1
w6 = w6 - eta * dE_total_net_o2 * out_h1
w7 = w7 - eta * dE_total_net_o1 * out_h2
w8 = w8 - eta * dE_total_net_o2 * out_h2

print(f"w1: {w1}")
print(f"w2: {w2}")
print(f"w3: {w3}")
print(f"w4: {w4}")
print(f"w5: {w5}")
print(f"w6: {w6}")
print(f"w7: {w7}")
print(f"w8: {w8}")
```
